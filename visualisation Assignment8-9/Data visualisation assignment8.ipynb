### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each datapoint with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition, datasets
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)
```

```python
fig = plt.figure(1, figsize=(4, 4))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, 2, 1], elev=25, azim=100)
plt.cla()
for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean()+3,
              X[y == label, 2].mean(), name,
              horizontalalignment='center')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
